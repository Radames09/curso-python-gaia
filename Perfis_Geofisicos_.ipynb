### Média

Podemos definir a **média** de um conjunto de observações $(\overline{x})$ da seguinte forma:
$$\overline{x} = \frac{1}{n}\sum_{i=1}^{n} x_i.$$

```python
# Funções de descrição estatística
df.mean()
```

### Mediana

A medida da média de uma série de atributos pode ser muito sensível à presença de *outliers*. Esse problema é minimizado utilizando a **mediana**.

Uma vez organizado em ordem crescente, a mediana de uma série pode ser definida como:

$$mediana(x) = \left\{\begin{matrix}
\frac{1}{2}\left ( x_{\frac{n}{2}} + x_{\frac{n}{2}+1} \right ), \ para \  n \ par\\ 
x_{\frac{n}{2}}, \ para \ n \ impar
\end{matrix}\right..$$

## Medidas de dispersão e distribuição

As medidas definidas em torno da média de um conjunto de valores são, em sua maioria, instanciações de uma medida denominada momento:

$$M_k(x) = \frac{\sum_{i=1}^{n} (x_i - \overline{x})^k}{(n-1)}$$

* 𝑘 = 2 é o segundo momento central, chamado de variância;
* 𝑘 = 3 é a obliquidade (skweness), ou terceiro momento central; e
* 𝑘 = 4 é a curtose (kurtosis), ou quarto momento central.

### Variância e desvio padrão

A variância (𝑣𝑎𝑟) é a medida mais utilizada para avaliar o espalhamento de valores:

$$M_2(x) = var = \frac{1}{n-1}\sum_{i=1}^{n} (x_i - \overline{x})^2$$.

É muito comum utilizar o desvio padrão:

$$\sigma = \sqrt{var}$$


```python
df.std()
```

```python
df.ILD.std()
```

### Obliquidade

A obliquidade (*skweness*), mede a simetria da distribuição dos dados em torno da média. Para tornar a medida independente de escala, é comum dividir o terceiro momento pelo desvio padrão ($\sigma$):

$$M_3(x) = \frac{1}{n-1} \frac{\sum_{i=1}^{n} (x_i - \overline{x})^3}{\sigma^3}.$$

* $𝑀_3$ = 0 indica que a distribuição analisada é aproximadamente simétrica;
* $𝑀_3$ > 0 (positiva) indica que a distribuição é assimétrica para o lado esquerdo;e 
* $𝑀_3$ < 0 (negativa) indica que a distribuição é assimétrica para o lado direito.

```python
df.skew()
```

### Curtose

Curtose (Kurtosis) é uma medida que captura o achatamento de uma determinada distribuição:

$$M_4(x) = \frac{1}{n-1} \frac{\sum_{i=1}^{n} (x_i - \overline{x})^4}{\sigma^4}.$$

* $𝑀_4$ = 0 a distribuição apresenta um achatamento próximo de uma normal;
* $𝑀_4$ > 0 a distribuição apresenta uma alta concentração de valores; e
* $𝑀_4$ < 0 a distribuição dos dados é mais achatada (parecendo-se mais com uma distribuição uniforme).

```python
df.kurtosis()
```

#### 3.2 Plot de perfis geofísicos na forma de track: Litologia | Resistividade | Perfis de Porosidade

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(10,15))
# Track Cáliper
axes[0].plot(df.CALI, df.DEPT, c='black', label='CALI') # plotagem do perfil
axes[0].set_title('Cáliper', fontsize=14) # adiciona título à pista
axes[0].set_xlabel(well.curves.CALI.unit.lower(), fontsize=14) # adiciona rótulo ao eixo x
axes[0].set_ylabel('Profundidade (m)', fontsize=14) # adiciona rótulo ao eixo y

#Track litológico
axes[1].plot(df.GR, df.DEPT, c='black', label='GR') # plotagem do perfil
axes[1].set_title('Raios Gama', fontsize=14) # adiciona título à pista
axes[1].set_xlabel(well.curves.GR.unit.lower(), fontsize=14) # adiciona rótulo ao eixo x

axes[2].plot(df.ILD, df.DEPT, c='black', label='ILD') # plotagem do perfil
axes[2].set_title('Resistividade', fontsize=14) # adiciona título à pista
axes[2].set_xlabel(well.curves.ILD.unit.lower(), fontsize=14) # adiciona rótulo ao eixo x

axes[3].plot(df.NPHI, df.DEPT, c='black', label='PHIN') # plotagem do perfil
#axes[3].set_title('Porosidade', fontsize=14)
#axes[3].set_xlabel(well.curves.NPHI.unit.lower(), fontsize=14)
axes[3].set_xlabel('$\phi_N$', fontsize=16) # adiciona rótulo ao eixo x
axes[3].set_xticks(np.arange(0, 1+0.2, 0.2)) # altera o "domínio da função"

ax3 = axes[3].twiny() # permite compartilhar o eixo y para duas plotagens distintas
ax3.plot(df.DPHI, df.DEPT, c='red', label='PHID',ls='--') # plotagem do perfil
ax3.set_xticks(np.arange(0, 1+0.2, 0.2)) # altera o "domínio da função"
ax3.invert_xaxis() #inverte o eixo y
ax3.set_xlabel('$\phi_D$', fontsize=16) # adiciona rótulo ao eixo x
ax3.legend() # habilita legenda

aux=0
for ax in axes:
    ax.grid() # habilita grid
    ax.invert_yaxis() # inverte eixos
    ax.legend() # habilita legenda
    if aux>0:
        ax.set_yticklabels([]) # desabilita números nos eixos 
    aux+=1
axes[3].legend(loc=2)
plt.tight_layout() # otimiza utilização de espaço
```
